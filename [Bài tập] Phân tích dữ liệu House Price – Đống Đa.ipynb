### 3. Lọc các thông tin Địa chỉ, Giá, Hướng nhà, Hướng ban công của các bản ghi có giấy chứng nhận sổ đỏ và có 3 phòng ngủ trở lên

```python
df2 = df[(df['land_certificate'] == 'Sổ đỏ') & (df['bedroom'] >= 3)].filter(['address','price','house_direction','balcony_direction'])
df2
```

### 4. Với mỗi loại nhà đất, tính trung bình cộng giá cũng như giá lớn nhất và giá nhỏ nhất

```python
a = df.groupby('type_of_land')['price'].mean()
b = df.groupby('type_of_land')['price'].min()
c = df.groupby('type_of_land')['price'].max()
print('Trung bình giá nhà', a)
print()
print('Giá nhà nhỏ nhất', b)
print()
print('Giá nhà lớn nhất', c)
```

### 5. Tính trung bình cộng số phòng ngủ, số phòng vệ sinh, số tầng của mỗi phường

```python
i = df.groupby('ward_name')['bedroom'].mean().round(1)
print('Số phòng ngủ trung bình',i)
print()
ii = df.groupby('ward_name')['toilet'].mean().round(1)
print('Số phòng ngủ trung bình',ii)
print()
iii = df.groupby('ward_name')['floor'].mean().round(1)
print('Số phòng ngủ trung bình',iii)
```
